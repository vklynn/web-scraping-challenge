# NASA mars News

```python
# Import Dependencies 
from splinter import Browser
import requests 
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager

import json
import pymongo
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
#URL for mars news to be scraped 
url = 'https://redplanetscience.com'
browser.visit(url)
```

```python
# Create a Beautiful Soup object
html = browser.html
parser = soup(browser.html,'html.parser')
table = parser.find_all('div',class_='list_text')
soup
```

```python
# Extract all article titles
mars_title = parser.select('div.list_text')
mars_title
```

```python
# Create an empty list to store titles and teaser 
mars_list = []
titles = []
descriptions = []
```

```python
# Loop through the text & extract title and teaser to store in dictionary 

for title in mars_title:
    titles.append(title.text)
    for description in mars_title:
        descriptions.append(description.text)
        mars_text_dict = {
            "content_title":title.text,
            "article_teaser_body": description.text
        }
        
        mars_list.append(mars_text_dict)
        
        
mars_dictionary = mars_list
```
